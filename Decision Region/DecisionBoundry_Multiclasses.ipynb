```python
from IPython.display import display

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
#
import sklearn
print("scikit-learn version: {}".format(sklearn.__version__))
#

#
############################# Modeling 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
############################# Modeling 
#

from sklearn.svm import LinearSVC

from sklearn.datasets import make_blobs
X, y = make_blobs(random_state=42)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0", "Class 1", "Class 2"])
#

'''
######################
# Linear SVM (1)     #
######################

linear_svm = LinearSVC().fit(X, y)
print("Coefficient shape: ", linear_svm.coef_.shape)
print("Intercept shape: ", linear_svm.intercept_.shape)

#
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_,['b', 'r', 'g']):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Line class 0', 'Line class 1','Line class 2'], loc=(1.01, 0.3))

#
mglearn.plots.plot_2d_classification(linear_svm, X, fill=True, alpha=.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_,['b', 'r', 'g']):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Line class 0', 'Line class 1','Line class 2'], loc=(1.01, 0.3))
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
#
'''
'''
######################
# Linear SVM (2)     #
######################

linear_svm = LinearSVC().fit(X, y)

mglearn.plots.plot_2d_classification(linear_svm, X, fill=True, alpha=.7)

plt.scatter(X[:, 0], X[:, 1], s=40, c=y, cmap=plt.cm.Spectral)
#plt.scatter(X0[:, 0], X0[:, 1], s=40, c=Y0)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
'''


######################
# KNN                #
######################

knn = KNeighborsClassifier(n_neighbors= 3).fit(X, y)

mglearn.plots.plot_2d_classification(knn, X, fill=True, alpha=.7)

plt.scatter(X[:, 0], X[:, 1], s=40, c=y, cmap=plt.cm.Spectral)
#plt.scatter(X0[:, 0], X0[:, 1], s=40, c=Y0)
plt.title("KNN")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()

'''
######################
# NN                 #
######################

mlp = MLPClassifier(solver='lbfgs', random_state=0).fit(X, y)
mlp = MLPClassifier(solver='lbfgs', activation='tanh',random_state=0, hidden_layer_sizes=[50,50]).fit(X, y)

mglearn.plots.plot_2d_classification(mlp, X, fill=True, alpha=.7)

#mglearn.discrete_scatter( X[:, 0], X[:, 1], y)
plt.scatter(X[:, 0], X[:, 1], s=40, c=y, cmap=plt.cm.Spectral)
plt.title("NN")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
'''

'''
######################
# RF                 #
######################

#forest = RandomForestClassifier(n_estimators=5,  random_state=2).fit(X, y)
#
forest = RandomForestClassifier(n_estimators=100,random_state=0, n_jobs=-1).fit(X, y)

mglearn.plots.plot_2d_classification(forest, X, fill=True, alpha=.7)

#mglearn.discrete_scatter( X[:, 0], X[:, 1], y)
plt.scatter(X[:, 0], X[:, 1], s=40, c=y, cmap=plt.cm.Spectral)
plt.title("RF")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
'''

######################
# RBF SVM            #
######################
#
svm = SVC(kernel='rbf', C=10, gamma=0.1).fit(X, y)
#svm = SVC(kernel='rbf', C=10, gamma=10).fit(X, Y)
#svm = SVC(kernel='rbf', C=3,  gamma=0.01).fit(X, y)

mglearn.plots.plot_2d_classification(svm, X, fill=True, alpha=.7)
plt.scatter(X[:, 0], X[:, 1], s=40, c=y, cmap=plt.cm.Spectral)
plt.title("RBF SVM")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()

```
